```python
import math
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
def f(x, y):
    return np.log(y**2 - 4*x + 8)

x = np.linspace(-3, 3, 100)  # 100 tevil med -3 in 3
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)  # mrea tok v ravnini xy
Z = f(X, Y)
fig = plt.figure(figsize=(12, 5))

# --- LEVO: 3D Graf ---
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax1.plot_surface(X, Y, Z, alpha=0.8)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# --- DESNO: Nivojnice ---
ax2 = fig.add_subplot(1, 2, 2)
levels = np.arange(-10, 10, 2)
contour = ax2.contour(X, Y, Z, levels=levels)
ax2.clabel(contour, inline=True, fontsize=8)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.grid(True)

plt.tight_layout()
```

```python
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.update_layout(
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    autosize=False,
    width=1000, height=800,
)
```
